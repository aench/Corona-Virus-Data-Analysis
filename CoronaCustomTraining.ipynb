```python
import pandas as pd
import numpy as np
import torch
from torch.nn import Linear
import torch.nn as nn
import seaborn as sns
import matplotlib as mlp
import matplotlib.pyplot as plt
import math
```

```python
dataFrameItalia = pd.read_csv("dati2.csv")
dataFrameItalia
```

```python
DataFrame_giorni = dataFrameItalia["NumeroGiorno"]
DataFrame_positivi = dataFrameItalia["Positivi"]

array = []
for i in range(0,59):
    array.append([DataFrame_giorni[i]])
tensor_prova = torch.Tensor(array)
tensor_giorni = tensor_prova

array=[]
for i in range(0,59):
    array.append([DataFrame_positivi[i]])
tensor_prova = torch.Tensor(array)
tensor_positivi = tensor_prova

print(tensor_giorni)
print(tensor_positivi)
```

```python
class Mnist_Logistic():
    def __init__(self):
        super().__init__()
        self.k = nn.Parameter(torch.tensor([0.1]), requires_grad = True)
        self.a = nn.Parameter(torch.tensor([10.0]), requires_grad = True)
        self.h = nn.Parameter(torch.tensor([300000.0]), requires_grad = True)
        self.b = nn.Parameter(torch.tensor([-100.0]), requires_grad = True)
        self.m = nn.Parameter(torch.tensor([-2000.0]), requires_grad = True)

    def getFirstParameter(self):
        return self.k
    
    def getSecondParameter(self):
        return self.a
    
    def getHeight(self):
        return self.h
    
    def getBias(self):
        return self.b
    
    def getLin(self):
        return self.m
    
    def forward(self, x):
        # print(a,b)
        pred = self.h/(1+self.a*(2**(-self.k*x)))+self.m*x+self.b
        #pred = self.linear(x)
        return pred

model = Mnist_Logistic()
a = model.getFirstParameter()
b = model.getSecondParameter()
print(a)
print(b)
tmp_list = []
for i in range(1,60):
    tmp_list.append([i])
print(tmp_list)
x = torch.Tensor(tmp_list)
print(x)
model.forward(x)
```

```python
def plotModel(title,model):
    plt.title = title
    k1 = model.getFirstParameter().detach().numpy()
    a1 = model.getSecondParameter().detach().numpy()
    h1 = model.getHeight().detach().numpy()
    b1 = model.getBias().detach().numpy()
    m1 = model.getLin().detach().numpy()
    print("Parameters: k=",k1," ; a=",a1," ; h=",h1," ; b=",b1," ; m=",m1)
    tmp_list = []
    for i in range(0,120):
        tmp_list.append(i)
    x1 = np.array(tmp_list)
    y1 = h1/(1+a1*(2**(-k1*x1)))+m1*x1+b1
    # print(y1)
    plt.plot(x1,y1, 'r')
    plt.plot(tensor_giorni,tensor_positivi, 'o')
    plt.show()
    
plotModel("Untrained Model",model)
```

```python
criterion = nn.MSELoss()
lrk = 0.000000000001
lra = 0.0000000001
lrh = 0.1
lrb = 0.00000001
lrm = 0.00001

#optimizer = torch.optim.SGD(model.parameters(), lr = 0.001) # stochastic gradient decent, minimizes one sample at a time, for big data better: computationally faster
epochs = 100
losses = []
for i in range(epochs):
    y_pred = model.forward(tensor_giorni)
    loss = criterion(y_pred, tensor_positivi)
    
    print('epoch:', i, 'loss:', loss.item())
    
    k = model.getFirstParameter()
    a = model.getSecondParameter()
    h = model.getHeight()
    b = model.getBias()
    m = model.getLin()
    
    #print(k,a,h,b)
    
    losses.append(loss)
    
    plotModel('Partially Trained Model',model)
    
    loss.backward()
    with torch.no_grad():
        k.sub_(a.grad*lrk)
        a.sub_(b.grad*lra)
        h.sub_(h.grad*lrh)
        b.sub_(b.grad*lrb)
        m.sub_(m.grad*lrm)
        k.grad.zero_()
        a.grad.zero_()
        h.grad.zero_()
        b.grad.zero_()
        m.grad.zero_()
```
