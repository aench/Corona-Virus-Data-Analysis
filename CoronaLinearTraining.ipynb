```python
DataFrame_giorni = dataFrameItalia["NumeroGiorno"]
DataFrame_positivi = dataFrameItalia["Positivi"]

array = []
for i in range(0,58):
    array.append([DataFrame_giorni[i]])
tensor_prova = torch.Tensor(array)
tensor_giorni = tensor_prova

array=[]
for i in range(0,58):
    array.append([DataFrame_positivi[i]])
tensor_prova = torch.Tensor(array)
tensor_positivi = tensor_prova

print(tensor_giorni)
print(tensor_positivi)
```

```python
# plt.plot(tensor_giorni,tensor_positivi, 'o')
```

```python
class LR(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size)
    def forward(self,x):
        pred = self.linear(x)
        return pred
model = LR(1,1)
[a, b] = model.parameters()
print(a,b)
```

```python
# dir(model)
```

```python
[w, b] = model.parameters()
print(w,b)
w1 = w[0][0].item()
b1 = b[0].item()
print(w1,b1)
def get_params():
  return (w[0][0].item(), b[0].item())
```

```python
def plot_fit(title):
    plt.title = title
    w1, b1 = get_params()
    print(w1,b1)
    x1 = np.array([0,58])
    y1 = w1*x1+b1
    plt.plot(x1,y1, 'r')
    plt.plot(tensor_giorni,tensor_positivi, 'o')
    plt.show()
```

```python
plot_fit("Initial model")
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.00001) # stochastic gradient decent, minimizes one sample at a time, for big data better: computationally faster
epochs = 100
losses = []
for i in range(epochs):
    y_pred = model.forward(tensor_giorni)
    loss = criterion(y_pred, tensor_positivi)
    
    print('epoch:', i, 'loss:', loss.item())
    [a , b] = model.parameters()
    a_num = a[0].item()
    b_num = b[0].item()
    print("Parameters: a=",a_num," ; b=",b_num)
    
    losses.append(loss)
    
    plot_fit('Partially Trained Model')
    
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plot_fit('Trained Model')
[w, b] = model.parameters()
print(w,b)
```
