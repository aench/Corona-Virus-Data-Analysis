# Logistic Model Training

```python
import pandas as pd
import numpy as np
import torch
from torch.nn import Linear
import torch.nn as nn
import seaborn as sns
import matplotlib as mlp
import matplotlib.pyplot as plt
import math
```

```python
dataFrameItalia = pd.read_csv("dati2.csv")
dataFrameItalia
```

```python
DataFrame_giorni = dataFrameItalia["NumeroGiorno"]
DataFrame_positivi = dataFrameItalia["Positivi"]

array = []
for i in range(0,59):
    array.append([DataFrame_giorni[i]])
tensor_prova = torch.Tensor(array)
tensor_giorni = tensor_prova

array=[]
for i in range(0,59):
    array.append([DataFrame_positivi[i]])
tensor_prova = torch.Tensor(array)
tensor_positivi = tensor_prova

print(tensor_giorni)
print(tensor_positivi)
```

```python
class Mnist_Logistic():
    def __init__(self):
        super().__init__()
        self.a = nn.Parameter(torch.tensor([0.1]), requires_grad = True)
        self.b = nn.Parameter(torch.tensor([1.0]), requires_grad = True)

    def getFirstParameter(self):
        return self.a
    
    def getSecondParameter(self):
        return self.b
    
    def forward(self, x):
        """
        [a,b] = self.parameters()
        a_num = a[0].item()
        b_num = b[0].item()
        tmp_list = []
        for i in x:
            tmp_list.append([i])
        x1 = torch.Tensor(tmp_list)
        tmp_list = []
        for j in x1:
            tmp_list.append([100000.0/(1+math.exp(-a_num*j))+b_num])
        tensor_prova = torch.Tensor(tmp_list)
        y_tensor = tensor_prova
        return y_tensor
        """
        # print(a,b)
        pred = 100000.0/(1+self.b*(2**(-self.a*x)))
        #pred = self.linear(x)
        return pred

model = Mnist_Logistic()
a = model.getFirstParameter()
b = model.getSecondParameter()
print(a)
print(b)
tmp_list = []
for i in range(1,60):
    tmp_list.append([i])
print(tmp_list)
x = torch.Tensor(tmp_list)
print(x)
model.forward(x)
```

```python
def plotModel(title,model):
    plt.title = title
    a1 = model.getFirstParameter().detach().numpy()
    b1 = model.getSecondParameter().detach().numpy()
    print(a1,b1)
    tmp_list = []
    for i in range(0,59):
        tmp_list.append(i)
    x1 = np.array(tmp_list)
    y1 = 100000.0/(1+b1*(2**(-a1*x1)))
    # print(y1)
    plt.plot(x1,y1, 'r')
    plt.plot(tensor_giorni,tensor_positivi, 'o')
    plt.show()
    
plotModel("Untrained Model",model)
```

```python
criterion = nn.MSELoss()
lr1 = 0.0000000001
lr2 = 0.000000000001
#optimizer = torch.optim.SGD(model.parameters(), lr = 0.001) # stochastic gradient decent, minimizes one sample at a time, for big data better: computationally faster
epochs = 10000
losses = []
for i in range(epochs):
    y_pred = model.forward(tensor_giorni)
    loss = criterion(y_pred, tensor_positivi)
    
    print('epoch:', i, 'loss:', loss.item())
    
    a = model.getFirstParameter()
    b = model.getSecondParameter()
    print(a,b)
    
    losses.append(loss)
    
    plotModel('Partially Trained Model',model)
    
    loss.backward()
    with torch.no_grad():
        a.sub_(a.grad*lr2)
        b.sub_(b.grad*lr1)
        a.grad.zero_()
        b.grad.zero_()
    
   
    
```

```python
plt.plot(range(epochs), losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
```

```python
plotModel('Trained Model', model)
```
